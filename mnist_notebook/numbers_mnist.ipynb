```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

#carrega os dados
data = tf.keras.datasets.mnist
(train_images,train_labels),(test_images,test_labels) = data.load_data()

#arquitetura do modelo
model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(),#camada que achata os dados, para ficarem na mesma dimensão das entradas
        tf.keras.layers.Dense(128,tf.nn.sigmoid),#camada intermediátira de processamento
        tf.keras.layers.Dense(10,tf.nn.softmax)#camada de saída com 10 classes de probabilidade
])

#compila o modelo
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])

#treinamento do modelo

history  = model.fit(train_images,train_labels,epochs=5)

model.evaluate(test_images,test_labels)


```

```python
history.history.keys()
plt.plot(history.history['accuracy'],label='Acurácia de Treinamento')
```

```python

pred = model.predict(test_images[0:1])
#print(pred)
predicted_class = np.argmax(pred)
print(predicted_class)
plt.imshow(test_images[0])
plt.show()
```
